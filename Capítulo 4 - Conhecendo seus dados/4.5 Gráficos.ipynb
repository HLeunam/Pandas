### 4.5 Gr√°ficos

```python
#P30 Gr√°fico de uma fun√ß√£o.
import pandas as pd

df = pd.DataFrame({"x": list(range(1,11))}) 
df['2024'] = (df.x * 2) + 3
lines = df.plot.line(x="x", y="2024", legend=True)
```

```python
#P31 - Gr√°fico de linhas.
dados = {
        '√≥leo de soja': [71.15, 66.93, 67.91, 67.66],
        'melancia': [10.52, 15.68, 14.95, 26.23],
        'frango congelado': [47.17, 23.68, 38.36, 17.08]
        }

df = pd.DataFrame(dados, index=[3,6,9,12])

lines = df.plot(kind ='line',
                style = ['.-', '+-', '*-'],
                ylim=(0,100), xlim=(3,12),
                title="Diferen√ßa de vendas ao longo dos meses"
                )
```

### Gr√°fico de barras

```python
# P32 Gr√°fico de barras vertical.
import pandas as pd

df = pd.DataFrame([70, 25, 50], index=['teatro', 'escultura', 'pintura'])

barras = df.plot(kind='bar', legend=True)
```

```python
#P33: Gr√°fico de barras agrupadas.
df = pd.DataFrame({'mulheres': [40, 10, 30],
                    'homens': [30,15,20]},
                    index=['teatro', 'escultura', 'pintura'])
barras = df.plot(kind='bar', legend=True, color=['pink', 'blue'])
```
